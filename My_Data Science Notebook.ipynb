```python
def countDog(st):
    count = 0
    for word in st.lower().split():
        if word == 'dog':
            count += 1
    return count
```

```python
def caught_speeding(speed, is_birthday):
    
    if is_birthday:
        speeding = speed - 5
    
    else:
        speeding = speed
         
    if speeding > 80 :
        return "bIG tICKET"
    elif speeding > 60:
        return "Small Ticket"
    else:
        return "No Ticket"
```

```python
caught_speeding(81,True)
```

```python
caught_speeding(81,False)
```

```python
def speedcal(speedlimit, my_birthday):
    if my_birthday:
        speeding = speedlimit - 5
    else:
        speeding = speedlimit
    if speeding > 80:
        return "Big Ticket"
    elif speeding > 60:
        return "Small Ticket"
    else:
        return "No Ticket"
```

```python
speedcal(81,True)
```

```python
speedcal(81,False)
```
